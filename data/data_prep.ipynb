```python
# imports
import os
import pandas as pd
from google.colab import drive
from rembg import remove
from sklearn.model_selection import train_test_split
```

```python
# mount google drive
drive.mount('/content/drive', force_remount=True)

# load image paths and create dataframe
def create_df(base_path):
    class_paths, classes = [], []
    for butterfly_dir in os.listdir(base_path):
        for img_path in os.listdir(f'{base_path}/{butterfly_dir}'):
            class_paths.append(f'{base_path}/{butterfly_dir}/{img_path}')
            classes.append(butterfly_dir)
    
    df = pd.DataFrame({'class_path': class_paths, 'class': classes})
    return df

# paths
base_path = '/content/drive/MyDrive/South FL Butterflies No Background'
df = create_df(base_path)
```

```python
# remove background
def remove_background(butterflies, start_index=0):
    for butterfly_dict in butterflies[start_index:]:
        input_path = butterfly_dict['path']
        file_name = input_path.split('/')[-1]
        
        if file_name == '.ipynb_checkpoints.jpeg':
            continue

        butterfly_directory = butterfly_dict['directory']
        output_dir = f'/content/drive/MyDrive/South FL Butterflies No Background/{butterfly_directory}'
        output_path = f'{output_dir}/{file_name}'

        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        with open(input_path, 'rb') as i, open(output_path, 'wb') as o:
            output = remove(i.read())
            o.write(output)
```

```python
# splitting data into train, test, and validation sets
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['class'])
train_df, valid_df = train_test_split(train_df, test_size=0.2, stratify=train_df['class'])
```
